```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
X,_ = make_blobs(n_samples=300, centers=3, random_state=4) #_ : place holder. To receive anything

X
```

```python
df_train = pd.DataFrame(data=X, columns=['x1','x2'])

df_train
```

```python
df_train.describe().T
```

```python
model = KMeans(n_clusters=2, random_state=2143)
model.fit(X)
new_data = np.array([[2,3],[0,4],[3,1]])
pred = model.predict(new_data)
```

```python
pred_all = model.predict(X)
pred_all
```

```python
df_train['cluster'] = pred_all
df_train
```

```python
sns.pairplot(data=df_train, hue='cluster')
plt.show()
```
